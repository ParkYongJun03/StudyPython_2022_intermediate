#### dataframe slicing
  - dataframe의 경우 기본적으로 [] 연산자가 **column 선택**에 사용
  - 하지만, slicing은 row 레벨로 지원

```python
train_data[0:5]
```

#### row 선택하기
  - Seires의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc, .iloc함수**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.info()
```

```python
import numpy as np
train_data.index =np.arange(100, 991)
```

```python
train_data.tail()
```

```python
train_data.loc[986]]
#표시된 인덱스로 따지는 loc
```

```python
train_data.iloc[890]
#실제 인덱스 위치를 따지는 iloc
```

```python
train_data.loc[[986, 100, 300]]
```

row, column 동시 선택

```python
train_data.loc[[986, 100, 110, 900], ['Survived', 'Name', 'Sex', 'Age']]
```

```python
train_data.loc[[986, 100, 110, 900], ['Survived', 'Sex', 'Name', 'Age']]
```
