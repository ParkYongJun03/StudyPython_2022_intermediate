```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://movie.naver.com/movie/running/current.naver'
resp = requests.get(url)
resp.text
soup = BeautifulSoup(resp.text)
```

```python
# select 메쏘드로 웹 크롤링하기
lst_dsc = soup.select("dl.lst_dsc")
#lst_dsc = soup.find_all('dl', class_='lst_dsc')
title_list=[]
score_list=[]
tot_num_list=[]
for dsc in lst_dsc:
    title = dsc.select_one("dt.tit a").text
    score = dsc.select_one("span.num").text
    info_star = dsc.find('dl', class_='info_star')
#    info_star_sel = dsc.select("dl.info_star") #얘는 왜 하위 속성 못 찾음?
    num = info_star.select("span.num2")
    num_sel =info_star.select("dl.info_star span.num2")
    tot_num=int;
    if len(num)==2 :
        a=num[0].select_one("em").text
        b=num[1].select_one("em").text
        temp1=a.split(',')
        temp2=b.split(',')
        a=''
        b=''
        for i in temp1:
            a=a+i
        for i in temp2:
            b=b+i
        tot_num=int(a)+int(b)
    else :
        a=num[0].select_one("em").text
        temp1=a.split(',')
        a=''
        for i in temp1:
            a=a+i
        tot_num=int(a)
    title_list.append(title)
    score_list.append(score)
    tot_num_list.append(tot_num)
```
